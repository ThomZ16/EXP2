```python
import numpy as np
import matplotlib.pyplot as plt
import netCDF4 as nc
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from datetime import datetime, timedelta
```

```python

# Fonction pour résoudre le système de Lorenz63
def lorenz63_solver(sigma, rho, beta, x0, y0, z0, dt=0.001, steps=20000):
    # Initialisation des tableaux pour stocker les résultats
    x = np.zeros(steps)
    y = np.zeros(steps)
    z = np.zeros(steps)

    # Conditions initiales
    x[0], y[0], z[0] = x0, y0, z0

    # Schéma d'Euler pour résoudre les équations
    for t in range(steps - 1):
        x[t + 1] = x[t] + sigma * (y[t] - x[t]) * dt
        y[t + 1] = y[t] + (x[t] * (rho - z[t]) - y[t]) * dt
        z[t + 1] = z[t] + (x[t] * y[t] - beta * z[t]) * dt

    return x, y, z

# Paramètres et conditions initiales
sigma = 10
rho = 1.0
beta = 2.667
eta = np.sqrt(beta * (rho - 1))

x0 = eta
y0 = eta + 3
z0 = rho - 1

# Nombre de pas de temps
steps = 20000
dt = 0.001

# Résoudre le système
x, y, z = lorenz63_solver(sigma, rho, beta, x0, y0, z0, dt, steps)

# Graphiques des solutions
time = np.linspace(0, steps * dt, steps)
plt.figure(figsize=(12, 8))

# x, y, z en fonction du temps
plt.subplot(3, 1, 1)
plt.plot(time, x, label='x(t)')
plt.xlabel('Temps')
plt.ylabel('x')
plt.xlim(0, time[-1])  # Assurer que l'axe des x commence bien à 0
plt.ylim(0, 5)
plt.title('Évolution de x en fonction du temps')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(time, y, label='y(t)')
plt.xlabel('Temps')
plt.ylabel('y')
plt.xlim(0, time[-1])  # Assurer que l'axe des x commence bien à 0
plt.ylim(0, 5)
plt.title('Évolution de y en fonction du temps')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(time, z, label='z(t)')
plt.xlabel('Temps')
plt.ylabel('z')
plt.xlim(0, time[-1])  # Assurer que l'axe des x commence bien à 0
plt.ylim(0, 5)
plt.title('Évolution de z en fonction du temps')
plt.legend()

plt.tight_layout()

# Sauvegarder la première figure
#plt.savefig(r'C:\Users\thoma\OneDrive\Images\Saved Pictures\expe2_fig1.png')
plt.show()

# Graphiques x vs y et x vs z
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x, y, label='x vs y')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0, 3)  # Limites de l'axe des x de 0 à 3
plt.ylim(0, 5)
plt.title('Graphique de x en fonction de y')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, z, label='x vs z')
plt.xlabel('x')
plt.ylabel('z')
plt.xlim(0, 3)  # Limites de l'axe des x de 0 à 3
plt.ylim(0, 5)
plt.title('Graphique de x en fonction de z')
plt.legend()

plt.tight_layout()

# Sauvegarder la deuxième figure
#plt.savefig(r'C:\Users\thoma\OneDrive\Images\Saved Pictures\expe2_fig2.png')
plt.show()
```

```python
# Paramètres du modèle de Lorenz
sigma = 10.0
rho = 28.0
beta = 8.0 / 3.0

def lorenz_system(state):
    x, y, z = state
    dx_dt = sigma * (y - x)
    dy_dt = x * (rho - z) - y
    dz_dt = x * y - beta * z
    return np.array([dx_dt, dy_dt, dz_dt])

# je suis passé ici à un schéma Runge kutta d'ordre 4, sans raison particulière, au départ je voulais voir si il y avait des différences
def rk4_step(f, state, dt):
    k1 = f(state)
    k2 = f(state + 0.5 * dt * k1)
    k3 = f(state + 0.5 * dt * k2)
    k4 = f(state + dt * k3)
    return state + (dt / 6.0) * (k1 + 2*k2 + 2*k3 + k4)

# Simulation du modèle de Lorenz
dt_lorenz = 0.01
N_lorenz = len(lat) * 100  # Augmentation du nombre de points pour une meilleure résolution
state0 = np.array([0.0, 1.0, 1.05])
state = state0.copy()
trajectory_lorenz = np.zeros((N_lorenz, 3))

for i in range(N_lorenz):
    trajectory_lorenz[i] = state
    state = rk4_step(lorenz_system, state, dt_lorenz)

x_lorenz = trajectory_lorenz[:, 0]
y_lorenz = trajectory_lorenz[:, 1]
```

```python
# Paramètres du modèle de Lorenz
sigma = 10.0
rho = 28.0
beta = 8.0 / 3.0

# Fonction pour le système de Lorenz
def lorenz_system(state):
    x, y, z = state
    dx_dt = sigma * (y - x)
    dy_dt = x * (rho - z) - y
    dz_dt = x * y - beta * z
    return np.array([dx_dt, dy_dt, dz_dt])

# Méthode de Runge-Kutta d'ordre 4 (RK4)
def rk4_step(f, state, dt):
    k1 = f(state)
    k2 = f(state + 0.5 * dt * k1)
    k3 = f(state + 0.5 * dt * k2)
    k4 = f(state + dt * k3)
    return state + (dt / 6.0) * (k1 + 2*k2 + 2*k3 + k4)

# Chemin vers le fichier NetCDF
file_path = r'C:\Users\thoma\Downloads\IBTrACS.ALL.v04r01.nc'

# Ouverture du fichier NetCDF
dataset = nc.Dataset(file_path, 'r')

# Accès aux variables
storm_names = nc.chartostring(dataset.variables['name'][:])
latitudes = dataset.variables['lat'][:]
longitudes = dataset.variables['lon'][:]
times = dataset.variables['time'][:]  # Le temps en jours depuis une date de référence
seasons = dataset.variables['season'][:]

# Extraire la date de référence depuis les attributs du temps (en jours)
time_units = dataset.variables['time'].units  # 'days since 1858-11-17 00:00:00'
time_base_str = time_units.split(' since ')[1]
time_base = datetime.strptime(time_base_str, '%Y-%m-%d %H:%M:%S')

# Fonction pour convertir le temps en dates lisibles (jours depuis time_base)
def convert_time(base_time, time_value):
    if np.ma.is_masked(time_value):  # Vérifie si la valeur est masquée
        return None  # Si le temps est une donnée masquée, on renvoie None
    return base_time + timedelta(days=time_value)

# Liste des indices des tempêtes 'MILTON'
milton_indices = [idx for idx, name in enumerate(storm_names) if name.strip().upper() == 'MILTON']

# Pour chaque tempête 'MILTON'
for idx in milton_indices:
    lat = latitudes[idx, :]
    lon = longitudes[idx, :]
    time = times[idx, :]
    season = seasons[idx]
    
    # Nettoyage des données manquantes
    valid_idx = ~np.isnan(lat) & ~np.isnan(lon) & ~np.ma.getmask(time)
    lat = lat[valid_idx]
    lon = lon[valid_idx]
    time = time[valid_idx]
    
    # Conversion des temps en dates
    dates = [convert_time(time_base, t) for t in time]
    
    # Supprimer les dates 'None' (provenant des données manquantes)
    dates = [d for d in dates if d is not None]
    
    # Simulation du modèle de Lorenz
    dt_lorenz = 0.01
    N_lorenz = len(lat) * 100
    state0 = np.array([0.0, 1.0, 1.05])  # Conditions initiales
    state = state0.copy()
    trajectory_lorenz = np.zeros((N_lorenz, 3))

    for i in range(N_lorenz):
        trajectory_lorenz[i] = state
        state = rk4_step(lorenz_system, state, dt_lorenz)

    x_lorenz = trajectory_lorenz[:, 0]
    y_lorenz = trajectory_lorenz[:, 1]

    # Normalisation des données du modèle de Lorenz
    x_norm = (x_lorenz - np.min(x_lorenz)) / (np.max(x_lorenz) - np.min(x_lorenz))
    y_norm = (y_lorenz - np.min(y_lorenz)) / (np.max(y_lorenz) - np.min(y_lorenz))

    lon_lorenz = np.min(lon) + x_norm * (np.max(lon) - np.min(lon))
    lat_lorenz = np.min(lat) + y_norm * (np.max(lat) - np.min(lat))

    # Création de la figure
    plt.figure(figsize=(10, 8))
    ax = plt.axes(projection=ccrs.PlateCarree())

    # Ajout des traits de côte et frontières
    ax.add_feature(cfeature.COASTLINE)
    ax.add_feature(cfeature.BORDERS)
    ax.add_feature(cfeature.LAND, facecolor='lightgray')
    ax.add_feature(cfeature.OCEAN, facecolor='lightblue')

    # Tracé de la trajectoire du cyclone Milton avec annotation des points clés
    ax.plot(lon, lat, 'r-', marker='o', transform=ccrs.PlateCarree(), label=f'Trajectoire de {storm_names[idx].strip()}')

    # Ajouter une annotation pour le premier et dernier point
    if dates:
        ax.text(lon[0], lat[0], dates[0].strftime('%Y-%m-%d'), color='red', fontsize=9, transform=ccrs.PlateCarree())
        ax.text(lon[-1], lat[-1], dates[-1].strftime('%Y-%m-%d'), color='red', fontsize=9, transform=ccrs.PlateCarree())
    
    # Tracé de la trajectoire du modèle de Lorenz
    ax.plot(lon_lorenz, lat_lorenz, 'b-', alpha=0.7, transform=ccrs.PlateCarree(), label='Trajectoire du modèle de Lorenz')



    # Ajout de la légende
    ax.legend()

    # Titre
    plt.title(f'Trajectoire du cyclone {storm_names[idx].strip()} et du modèle de Lorenz (Saison {season})')
    plt.savefig(r'C:\Users\thoma\OneDrive\Images\Saved Pictures\expe2_figMilton.png')


    # Affichage de la carte
    plt.show()
```
